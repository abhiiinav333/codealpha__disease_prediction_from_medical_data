```python
disease = ['Fungal infection','Allergy','GERD','Chronic cholestasis','Drug Reaction',
        'Peptic ulcer diseae','AIDS','Diabetes ','Gastroenteritis','Bronchial Asthma','Hypertension ',
        'Migraine','Cervical spondylosis',
        'Paralysis (brain hemorrhage)','Jaundice','Malaria','Chicken pox','Dengue','Typhoid','hepatitis A',
        'Hepatitis B','Hepatitis C','Hepatitis D','Hepatitis E','Alcoholic hepatitis','Tuberculosis',
        'Common Cold','Pneumonia','Dimorphic hemmorhoids(piles)',
        'Heart attack','Varicose veins','Hypothyroidism','Hyperthyroidism','Hypoglycemia','Osteoarthristis',
        'Arthritis','(vertigo) Paroymsal  Positional Vertigo','Acne','Urinary tract infection','Psoriasis',
        'Impetigo']
```

```python
prognosis_list = list(trained['prognosis'])
```

```python
new_prognosis_list = []
for value in prognosis_list:
    index = disease.index(value)# Find the index of the disease in the `disease` list
    new_prognosis_list.append(index)  # Append the index to the new list
```

```python
trained['prognosis'] = new_prognosis_list
```

```python
prognosis_test = list(tested['prognosis'])  # Convert prognosis column to a list
```

```python
new_prognosis_test = []
for value in prognosis_test:
    index = disease.index(value)  # Find the index of the disease in the `disease` list
    new_prognosis_test.append(index)  # Append the index to the new list
tested['prognosis'] = new_prognosis_test
```

```python
xtrain = trained[symptoms]
ytrain = trained['prognosis']

xtest = tested[symptoms]
ytest = tested['prognosis']
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
model = MultinomialNB().fit(xtrain, ytrain)
# Test the model
ypred = model.predict(xtest)
accuracy = accuracy_score(ytest, ypred)
print("Model Accuracy is :", accuracy)
```

```python
# Function to predict disease from symptoms
def diseased(symptom):
    sympt_map = [] 
    for i in symptoms:
        # If the symptom is in the user's list, append 1, otherwise append 0
        if i in symptom:
            sympt_map.append(1)
        else:
            sympt_map.append(0)
    # print(input_vector)
    prediction = model.predict([sympt_map])[0]
    return disease[prediction]
```

```python
example_case = ['chest_pain', 'bloody_stool', 'nausea','vomiting','cough'] 
predicted_disease = diseased(example_case)
print('Predicted Disease is :',predicted_disease)
```
